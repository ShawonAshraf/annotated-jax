```python
import jax
import jax.numpy as jnp
```

```python
key = jax.random.PRNGKey(0)
```

```python
x = jax.random.normal(key, (10, 5))
```

```python
y = jax.random.normal(key, (10, 5))
```

```python
res = jax.vmap(jnp.dot)
```

```python
res(x, y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

a = jnp.abs(jax.random.normal(key, (1,)))
b = jax.random.randint(key=key, shape=(1,), minval=0, maxval=100)

f = lambda x: a*x + b

x = jax.random.randint(key=key, shape=(100, ), minval=0, maxval=100)
y = f(x)

plt.plot(x, y)
plt.show()
```
